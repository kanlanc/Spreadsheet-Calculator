## Method

### 1. Create a dataframe with inputs evaluating pure expressions
Take the inputs and evaluate the ones that only have numeric values and operators in them first. This can be done using a stack and is further explained in the functions section

Doing so, will help populate the dataframe thus preventing None error or returning 0 when asked for a particular cell that hasn't gotten it's input yet.

**Personal Note: Invalid expressions are not being checked for**


### 2. Loop the dataframe again and evaluate the cell expressions

In this situation to reduce unnecessary loops, I am directly printing the result while I evaluate any linked cells expressions while changing the CSV.

For better readibility, two loops could exist where one evaluates the expressions and the other goes row wise printing 
results

Finally, Format the cells in print to give 5 integers in float

A single function could be created for both of them but that would be bad for readability so two different functions for pure expressions and linked cells expressions have been created
